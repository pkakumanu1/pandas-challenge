```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```


## Player Count


* Display the total number of players


```python
Total_Players = str(len(purchase_data["SN"].unique()))
TotalPlayer_df = pd.DataFrame([{"Total Players " : Total_Players}])
TotalPlayer_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Unique_items = len(purchase_data["Item ID"].unique())
Avg_price = '${:,.2f}'.format(purchase_data["Price"].mean())
NumberOfPurchases = purchase_data["Purchase ID"].count()
TotalRevenue = '${:,.2f}'.format(purchase_data["Price"].sum())
Summary_df = pd.DataFrame([{"Number of Unique Items": Unique_items,
                           "Average Price" : Avg_price,
                           "Number of Purchases" : NumberOfPurchases,
                           "Total Revenue" : TotalRevenue}])
Summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
Gender_df = purchase_data[["SN","Gender"]].drop_duplicates()
GroupedGender = Gender_df.groupby("Gender").count().rename(columns={"SN":"Total Count"}).sort_values("Total Count",ascending=False)
GroupedGender["Percentage of Players"] = ((GroupedGender["Total Count"]/float(Total_Players))*100).map('{:,.2f}%'.format)
GroupedGender
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python


GroupedPurchaseData = purchase_data.groupby("Gender")
Group_Purchase_df = pd.DataFrame([GroupedPurchaseData["Purchase ID"].count().rename("Purchase Count"),
                                  GroupedPurchaseData["Price"].mean().rename("Average Purchase Price"),
                                  GroupedPurchaseData["Price"].sum().rename("Total Purchase Value")
                                 ])

Group_Purchase_df = Group_Purchase_df.transpose()
Group_Purchase_df["Avg Total Purchase per Person"] = Group_Purchase_df["Total Purchase Value"]/GroupedGender["Total Count"]

Group_Purchase_df["Purchase Count"] = Group_Purchase_df["Purchase Count"].map("{:,.0f}".format)
Group_Purchase_df["Average Purchase Price"] = Group_Purchase_df["Average Purchase Price"].map("${:,.2f}".format)
Group_Purchase_df["Total Purchase Value"] = Group_Purchase_df["Total Purchase Value"].map("${:,.2f}".format)
Group_Purchase_df["Avg Total Purchase per Person"] = Group_Purchase_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
Group_Purchase_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#purchase_data["Age"].max()
bins = [0,9,14,19,24,29,34,39,45]
group_names = [" <10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
AgeGroup = pd.cut(purchase_data["Age"],bins,labels=group_names,include_lowest=True)
AgeGroup.head()
purchase_data["AgeGroup"]= pd.DataFrame(AgeGroup)
purchase_data.head()
Age_Demo = purchase_data.groupby("AgeGroup")["SN"].nunique()
Age_Demo_df = pd.DataFrame(Age_Demo).rename(columns={"SN":"Total Count"})
Age_Demo_df["Percentage of Players"] = ((Age_Demo_df["Total Count"]/float(Total_Players))*100).map('{:,.2f}%'.format)

Age_Demo_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data.head()
GroupPurchase = purchase_data.groupby("AgeGroup")
GroupPurchase_df = pd.DataFrame([GroupPurchase["Purchase ID"].count().rename("Purchase count"),
                                GroupPurchase["Price"].mean().rename("Average Purchase Price"),
                                GroupPurchase["Price"].sum().rename("Total Purchase Value")])
GroupPurchase_df = GroupPurchase_df.transpose()
GroupPurchase_df["Avg Total Purchase per Person"] = GroupPurchase_df["Total Purchase Value"]/GroupPurchase_df["Purchase count"]

GroupPurchase_df["Purchase count"] = GroupPurchase_df["Purchase count"].map('{:,.0f}'.format)
GroupPurchase_df["Average Purchase Price"] = GroupPurchase_df["Average Purchase Price"].map('${:,.2f}'.format)
GroupPurchase_df["Total Purchase Value"] = GroupPurchase_df["Total Purchase Value"].map('${:,.2f}'.format)
GroupPurchase_df["Avg Total Purchase per Person"] = GroupPurchase_df["Avg Total Purchase per Person"].map('${:,.2f}'.format)
GroupPurchase_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
GroupSN = purchase_data.groupby("SN").agg({"Item ID" : ['count'],"Price" :['mean', 'sum']}).rename(columns={})
GroupSN.columns =GroupSN.columns.droplevel(0)


GroupSN = GroupSN.rename(columns={"count": "Purchase count", 
                                             "mean": "Average Purchase Price",
                                             "sum":"Total Purchase Value"})




GroupSN.reset_index()

GroupSN=GroupSN.sort_values("Total Purchase Value",ascending=False)


GroupSN["Average Purchase Price"] =GroupSN["Average Purchase Price"].map('${:,.2f}'.format)
GroupSN["Total Purchase Value"] =GroupSN["Total Purchase Value"].map('${:,.2f}'.format)

GroupSN.head()
```
